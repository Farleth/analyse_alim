```python
def info_sous_nutr(year:int):
    
    pop_sous_nutr = df_sous_nutr.loc[df_sous_nutr['Ann√©e'] == f"{year-1}-{year+1}"].Valeur.sum()
    print(f"{pop_sous_nutr} Millions d'habitants en sous-nutrition en {year} \n ")
    pop_total_year = pop_by_year.loc[pop_by_year['Ann√©e'] == year].Valeur / 1000
    pct_pop_sous_nutr = 100 * pop_sous_nutr / (pop_by_year.loc[pop_by_year['Ann√©e'] == year].Valeur / 1000)
    print(f"{pct_pop_sous_nutr.round(2)}% de la pop. mondial en sous nutrition. \n")
    pop_year = df_pop.loc[df_pop["Ann√©e"] == year]
    dispo_by_zone_and_pop = dispo_by_zone.merge(pop_year, on="Zone", how="left")
    dispo_by_zone_and_pop["population"] = dispo_by_zone_and_pop["Valeur"]
    dispo_by_zone_and_pop = dispo_by_zone_and_pop.drop(columns="Valeur")
    dispo_by_zone_and_pop["dispo_alim"] = dispo_by_zone_and_pop["Disponibilit√© alimentaire (Kcal/personne/jour)"] * dispo_by_zone_and_pop["population"] / 2500
    print(f'{(dispo_by_zone_and_pop.dispo_alim.sum() * 100) / dispo_by_zone_and_pop["population"].sum()}% des √™tre humains pouvant th√©oriquement √™tre nourris \n')
    dispo_by_origin_and_pop = dispo_by_origin.merge(pop_year, on="Zone", how="left")
    dispo_by_origin_and_pop["dispo_alim"] = dispo_by_origin_and_pop["Disponibilit√© alimentaire (Kcal/personne/jour)"] * dispo_by_origin_and_pop["Valeur"] / 2500
    print(f'{(dispo_by_origin_and_pop.dispo_alim.sum() * 100) / dispo_by_origin_and_pop["Valeur"].sum()}% des √™tre humains pouvant th√©oriquement √™tre nourris seulement par la production v√©g√©tale \n')
    pop_sous_nutr_by_zone = df_sous_nutr.loc[df_sous_nutr['Ann√©e'] == f"{year-1}-{year+1}"]
    pop_by_year_zone = df_pop.groupby(by=["Ann√©e", "Zone"]).sum()
    pop_by_year_zone = pop_by_year_zone.reset_index()
    pop_by_year_zone = pop_by_year_zone.merge(pop_sous_nutr_by_zone, on="Zone", how="left")
    pop_by_year_zone = pop_by_year_zone.loc[pop_by_year_zone.Ann√©e_x == year].dropna()
    pop_by_year_zone.rename(columns={"Valeur_x" : "population", "Valeur_y" : "sous_nutr(Mhab)"}, inplace=True)
    pop_by_year_zone.drop(columns="Ann√©e_y", inplace=True)
    pop_by_year_zone["population"] = pop_by_year_zone["population"] / 1000
    pop_by_year_zone.rename(columns={"population" : "pop(Mhab)"}, inplace=True)
    pop_by_year_zone["pct_sous_nutr"] = pop_by_year_zone["sous_nutr(Mhab)"] * 100 / pop_by_year_zone["pop(Mhab)"]
    dispo_by_zone_and_pop["pct_dispo_alim/hab"] = dispo_by_zone_and_pop.dispo_alim * 100 / dispo_by_zone_and_pop.population
    h = pop_by_year_zone.sort_values(by="pct_sous_nutr", ascending=False).head(10)
    print(h)
```
